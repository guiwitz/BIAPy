```python
import numpy as np
import skimage
import skimage.io
import skimage.measure
import matplotlib.pyplot as plt
```

## Import image

```python
image = skimage.io.imread('../Data/BBBC007_v1_images/A9/A9 p10d.tif')
```

## Threshold the image

Define a theshold value:

```python
threshold = 100
```

```python
mask = image > threshold
```

## Plot the image

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(image, cmap = 'gray')
ax[1].imshow(mask, cmap = 'gray')
```

## Analyze the result

```python
sizes = skimage.measure.regionprops_table(skimage.measure.label(mask), properties=('label', 'area'))
```

```python
plt.hist(sizes['area'], np.arange(0,200,20));
```
